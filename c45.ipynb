```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mhs = pd.read_excel("train.xlsx" )
mhs.head(10)
```

```python
x = mhs.drop(["STATUS KELULUSAN"], axis=1)
x.head(5)

#x = data[['STATUS MAHASISWA','STATUS NIKAH','IPS 1','IPS 2','IPS 3','IPS 4','IPS 5','IPS 6','IPS 7','IPS 8','IPK']]
```

```python
y = mhs["STATUS KELULUSAN"]
y.head(10)
```

```python
# import c4.5
from sklearn.tree import DecisionTreeClassifier
```

```python
#c45 model
model = DecisionTreeClassifier()
```

```python
#memasukkan data training pada fungsi c45   
data_training = model.fit(x,y)
```

```python
#model dalam bentuk pickle
import pickle
with open('model.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
#Melakukan prediksi pada data mining
y_predict = data_training.predict(x)
```

```python
mahasiswa = pd.read_excel("test.xlsx") 
mahasiswa.head(10)               
```

```python
xtest = mahasiswa.drop(["STATUS KELULUSAN"], axis=1)
xtest.head(5)
ytest_predict = data_training.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
ytest = mahasiswa["STATUS KELULUSAN"]       
print("Nilai Akurasi = %0.2f" % accuracy_score(ytest,ytest_predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ytest_predict))
```
