```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("train.xlsx") 

data.fillna(0, inplace=True)  # Untuk mengganti NaN menjadi 0
data.replace("NaN", '0', inplace=True)
```

```python
data.head(300)
```

```python
x = data.drop(["STATUS KELULUSAN"], axis=1)
x.head(5)
```

```python
y = data["STATUS KELULUSAN"]
y.head(10)
```

```python
#import knn 
from sklearn.neighbors import KNeighborsClassifier
```

```python
#model knn
model = KNeighborsClassifier(n_neighbors=5)
```

```python
#memasukkan data training pada fungsi knn
data_training = model.fit(x,y)
```

```python
#model dalam bentuk pickle
import pickle
with open('knn.pkl', 'wb') as f:
    pickle.dump(model, f)
```

```python
#Melakukan prediksi pada data mining
y_predict = data_training.predict(x)
```

```python
mahasiswa = pd.read_excel("test.xlsx" )
mahasiswa.fillna(0, inplace=True)  # Untuk mengganti NaN menjadi 0
mahasiswa.replace("NaN", '0', inplace=True)
mahasiswa.head(10)   
```

```python
xtest = mahasiswa.drop(["STATUS KELULUSAN"], axis=1)
xtest.head(5)
ytest_predict = data_training.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
ytest = mahasiswa["STATUS KELULUSAN"]       
print("Nilai Akurasi = %0.2f" % accuracy_score(ytest,ytest_predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ytest_predict))
```
